# Load the dataframe

```python
import pandas as pd

# Define the file path
data_path = '/Users/sofianewalid/Desktop/datassubs.xlsx'

# Load the DataFrame
df = pd.read_excel(data_path)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# Display the properties of the DataFrame
print(df.info())
print(df.describe())
```

```python
import pandas as pd
import re

# Create a dictionary to store sub-dataframes for each row
sub_dfs = {}

# Iterate over each row in the main DataFrame
for index, row in df.iterrows():
    author_pub_history = row['authorPublicationHistory']

    # Confirm it is a string
    if isinstance(author_pub_history, str):
        publications_data = []

        # Use regex to find all dictionary-like entries in the string
        dict_matches = re.findall(r"\{.*?\}", author_pub_history)

        for match in dict_matches:
            publication_entry = {
                'title': None,
                'year': None,
                'abstract': None,
                'doi': None,
                'citation_count': None,
                'scopus_paper_id': None,
                'Authors': None
            }

            # Extract values for each key and handle the 'Abstract' field, with support for single and double quotes
            publication_entry['title'] = re.search(r"'title':\s*'([^']*)'", match)
            publication_entry['year'] = re.search(r"'year':\s*'([^']*)'", match)
            publication_entry['abstract'] = re.search(r"'Abstract':\s*['\"]([^'\"]*)['\"]", match)  # Support for both single and double quotes
            publication_entry['doi'] = re.search(r"'doi':\s*'([^']*)'", match)
            publication_entry['citation_count'] = re.search(r"'citation_count':\s*'([^']*)'", match)
            publication_entry['scopus_paper_id'] = re.search(r"'scopus_paper_id':\s*'([^']*)'", match)
            publication_entry['Authors'] = re.search(r"'Authors':\s*\[(.*?)\]", match)
            # Replace the match object with the actual string if found, else keep as None
            for key, value in publication_entry.items():
                if value:
                    publication_entry[key] = value.group(1)

            # Append the publication entry to the list
            publications_data.append(publication_entry)

        # Create a DataFrame from the collected data and store it in the dictionary
        publications_df = pd.DataFrame(publications_data)
        sub_dfs[index] = publications_df

# Print confirmation of sub-dataframes creation
print("Sub-dataframes for each row have been created.")

# Example: Access the sub-dataframe for row 0
if 0 in sub_dfs:
    print("Accessing the sub-dataframe for row 0:")
    display(sub_dfs[0])
```

# Adding the IsMatch column

```python
# Iterate through each key-value pair in the dictionary
for index, publications_df in sub_dfs.items():
    # Get the DOI of the corresponding row in the main DataFrame
    original_doi = df.loc[index, 'doi']
    
    # Add the 'IsMatch' column: 1 if the DOI matches, 0 otherwise
    publications_df['IsMatch'] = publications_df['doi'].apply(lambda x: 1 if x == original_doi else 0)

# Print confirmation
print("The 'IsMatch' column has been added to each sub-DataFrame.")

# Example: Display the sub-DataFrame for row 0 to verify
if 0 in sub_dfs:
    print("Sub-DataFrame for row 0 with 'IsMatch' column:")
    display(sub_dfs[0])
```

# Calling and loading the embeddings

```python
# Import necessary libraries
from sentence_transformers import SentenceTransformer
import tensorflow_hub as hub
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load models
print("Loading models...")
sbert_model = SentenceTransformer('all-MiniLM-L6-v2')
use_model = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
print("Models loaded successfully.")

# Compute embeddings and similarities directly on pre-existing sub_dfs
print("Computing embeddings and similarities...")
for index, publications_df in sub_dfs.items():
    # Extract submission title and abstract for the current row in the main DataFrame
    submission_title = df.loc[index, 'submissionTitle']
    submission_abstract = df.loc[index, 'submissionAbstract']

    # Compute embeddings for the submission title and abstract
    sbert_title_embedding = sbert_model.encode(submission_title, convert_to_tensor=True)
    sbert_abstract_embedding = sbert_model.encode(submission_abstract, convert_to_tensor=True)
    use_title_embedding = use_model([submission_title])[0].numpy()
    use_abstract_embedding = use_model([submission_abstract])[0].numpy()

    # Initialize lists to store similarities
    title_similarities = {'sbert': [], 'use': [], 'tfidf': []}
    abstract_similarities = {'sbert': [], 'use': [], 'tfidf': []}

    # Iterate through each row in the sub-DataFrame
    for _, row in publications_df.iterrows():
        title = row['title'] if row['title'] else ""
        abstract = row['abstract'] if row['abstract'] else ""

        # Compute S-BERT and USE embeddings
        sbert_title = sbert_model.encode(title, convert_to_tensor=True)
        sbert_abstract = sbert_model.encode(abstract, convert_to_tensor=True)
        use_title = use_model([title])[0].numpy()
        use_abstract = use_model([abstract])[0].numpy()

        # Compute S-BERT and USE similarities
        title_similarities['sbert'].append(cosine_similarity([sbert_title_embedding.cpu().numpy()], [sbert_title.cpu().numpy()])[0][0])
        abstract_similarities['sbert'].append(cosine_similarity([sbert_abstract_embedding.cpu().numpy()], [sbert_abstract.cpu().numpy()])[0][0])
        title_similarities['use'].append(cosine_similarity([use_title_embedding], [use_title])[0][0])
        abstract_similarities['use'].append(cosine_similarity([use_abstract_embedding], [use_abstract])[0][0])

        # Compute TF-IDF similarities
        titles = [submission_title, title]
        abstracts = [submission_abstract, abstract]
        tfidf_title_vectors = TfidfVectorizer().fit_transform(titles)
        tfidf_abstract_vectors = TfidfVectorizer().fit_transform(abstracts)
        title_similarities['tfidf'].append(cosine_similarity(tfidf_title_vectors[0], tfidf_title_vectors[1])[0][0])
        abstract_similarities['tfidf'].append(cosine_similarity(tfidf_abstract_vectors[0], tfidf_abstract_vectors[1])[0][0])

    # Add computed similarities to the sub-DataFrame
    publications_df['s-bert title similarity'] = title_similarities['sbert']
    publications_df['s-bert abstract similarity'] = abstract_similarities['sbert']
    publications_df['use title similarity'] = title_similarities['use']
    publications_df['use abstract similarity'] = abstract_similarities['use']
    publications_df['tfidf title similarity'] = title_similarities['tfidf']
    publications_df['tfidf abstract similarity'] = abstract_similarities['tfidf']

print("Embeddings and similarities computed.")

# Example: Display all sub-DataFrames
for key, sub_df in sub_dfs.items():
    print(f"Sub-DataFrame for row {key} with similarity columns:")
    display(sub_df)
```

```python
# Compute embeddings and similarities with additional aggregated measures
print("Computing embeddings, similarities, and aggregated measures...")
for index, publications_df in sub_dfs.items():
    # Extract submission title and abstract for the current row in the main DataFrame
    submission_title = df.loc[index, 'submissionTitle']
    submission_abstract = df.loc[index, 'submissionAbstract']

    # Compute embeddings for the submission title and abstract
    sbert_title_embedding = sbert_model.encode(submission_title, convert_to_tensor=True)
    sbert_abstract_embedding = sbert_model.encode(submission_abstract, convert_to_tensor=True)
    use_title_embedding = use_model([submission_title])[0].numpy()
    use_abstract_embedding = use_model([submission_abstract])[0].numpy()

    # Initialize lists to store similarities
    title_similarities = {'sbert': [], 'use': [], 'tfidf': []}
    abstract_similarities = {'sbert': [], 'use': [], 'tfidf': []}

    for _, row in publications_df.iterrows():
        title = row['title'] if row['title'] else ""
        abstract = row['abstract'] if row['abstract'] else ""

        # Compute S-BERT and USE embeddings
        sbert_title = sbert_model.encode(title, convert_to_tensor=True)
        sbert_abstract = sbert_model.encode(abstract, convert_to_tensor=True)
        use_title = use_model([title])[0].numpy()
        use_abstract = use_model([abstract])[0].numpy()

        # Compute S-BERT and USE similarities
        title_similarities['sbert'].append(cosine_similarity([sbert_title_embedding.cpu().numpy()], [sbert_title.cpu().numpy()])[0][0])
        abstract_similarities['sbert'].append(cosine_similarity([sbert_abstract_embedding.cpu().numpy()], [sbert_abstract.cpu().numpy()])[0][0])
        title_similarities['use'].append(cosine_similarity([use_title_embedding], [use_title])[0][0])
        abstract_similarities['use'].append(cosine_similarity([use_abstract_embedding], [use_abstract])[0][0])

        # Compute TF-IDF similarities
        titles = [submission_title, title]
        abstracts = [submission_abstract, abstract]
        tfidf_title_vectors = TfidfVectorizer().fit_transform(titles)
        tfidf_abstract_vectors = TfidfVectorizer().fit_transform(abstracts)
        title_similarities['tfidf'].append(cosine_similarity(tfidf_title_vectors[0], tfidf_title_vectors[1])[0][0])
        abstract_similarities['tfidf'].append(cosine_similarity(tfidf_abstract_vectors[0], tfidf_abstract_vectors[1])[0][0])

    # Add computed similarities to the sub-DataFrame
    publications_df['s-bert title similarity'] = title_similarities['sbert']
    publications_df['s-bert abstract similarity'] = abstract_similarities['sbert']
    publications_df['use title similarity'] = title_similarities['use']
    publications_df['use abstract similarity'] = abstract_similarities['use']
    publications_df['tfidf title similarity'] = title_similarities['tfidf']
    publications_df['tfidf abstract similarity'] = abstract_similarities['tfidf']

    # Add aggregated measures for S-BERT
    publications_df['s-bert_1'] = (publications_df['s-bert title similarity'] + publications_df['s-bert abstract similarity']) / 2
    publications_df['s-bert_2'] = (0.3 * publications_df['s-bert title similarity'] + 0.7 * publications_df['s-bert abstract similarity'])
    publications_df['s-bert_3'] = (0.7 * publications_df['s-bert title similarity'] + 0.3 * publications_df['s-bert abstract similarity'])
    publications_df['s-bert_4'] = publications_df['s-bert title similarity']  # Only title similarity

    # Add aggregated measures for USE
    publications_df['use_1'] = (publications_df['use title similarity'] + publications_df['use abstract similarity']) / 2
    publications_df['use_2'] = (0.3 * publications_df['use title similarity'] + 0.7 * publications_df['use abstract similarity'])
    publications_df['use_3'] = (0.7 * publications_df['use title similarity'] + 0.3 * publications_df['use abstract similarity'])
    publications_df['use_4'] = publications_df['use title similarity']  # Only title similarity

    # Add aggregated measures for TF-IDF
    publications_df['tfidf_1'] = (publications_df['tfidf title similarity'] + publications_df['tfidf abstract similarity']) / 2
    publications_df['tfidf_2'] = (0.3 * publications_df['tfidf title similarity'] + 0.7 * publications_df['tfidf abstract similarity'])
    publications_df['tfidf_3'] = (0.7 * publications_df['tfidf title similarity'] + 0.3 * publications_df['tfidf abstract similarity'])
    publications_df['tfidf_4'] = publications_df['tfidf title similarity']  # Only title similarity

print("Embeddings, similarities, and aggregated measures computed.")

# Example: Display all sub-DataFrames
for key, sub_df in sub_dfs.items():
    print(f"Sub-DataFrame for row {key} with similarity and aggregated columns:")
    display(sub_df)
```

```python
# Initialize a dictionary to store bi-accuracy results for all models and aggregation types
bi_accuracy_results = {
    's-bert_1': 0, 's-bert_2': 0, 's-bert_3': 0, 's-bert_4': 0,
    'use_1': 0, 'use_2': 0, 'use_3': 0, 'use_4': 0,
    'tfidf_1': 0, 'tfidf_2': 0, 'tfidf_3': 0, 'tfidf_4': 0,
}

# Initialize a list to store detailed results for each row
bi_accuracy_detailed_results = []

# Keep track of the number of valid sub-DataFrames (those with at least one `IsMatch = 1`)
valid_sub_dfs_count = 0

# Compute bi-accuracy for each aggregation measure
print("Checking bi-accuracy for each row and column...")
for index, publications_df in sub_dfs.items():
    # Filter rows where IsMatch is 1
    matching_rows = publications_df[publications_df['IsMatch'] == 1]

    # Skip this sub-DataFrame if there are no `IsMatch = 1` rows
    if len(matching_rows) == 0:
        continue

    # Increment the count of valid sub-DataFrames
    valid_sub_dfs_count += 1

    # For each aggregated column, check bi-accuracy for the current sub-DataFrame
    for agg_column in ['s-bert_1', 's-bert_2', 's-bert_3', 's-bert_4',
                       'use_1', 'use_2', 'use_3', 'use_4',
                       'tfidf_1', 'tfidf_2', 'tfidf_3', 'tfidf_4']:
        # Get the indices of the top 2 rows for the current aggregated similarity column
        top_2_indices = publications_df.nlargest(2, agg_column).index

        # Check if any of the matching rows are in the top 2
        is_good_match = any(matching_rows.index.isin(top_2_indices))

        # Update the bi-accuracy results
        if is_good_match:
            bi_accuracy_results[agg_column] += 1

        # Store the detailed result for this row and column
        bi_accuracy_detailed_results.append({
            'row_index': index,
            'column': agg_column,
            'is_good_match': is_good_match
        })

# Normalize bi-accuracy by dividing by the total number of valid sub-DataFrames
for key in bi_accuracy_results:
    bi_accuracy_results[key] /= valid_sub_dfs_count

# Print the overall bi-accuracy results
print("\nBi-Accuracy Results:")
for model, bi_acc in bi_accuracy_results.items():
    print(f"{model}: {bi_acc:.2f}")

# Print detailed results for every row
print("\nDetailed Bi-Accuracy Results:")
for result in bi_accuracy_detailed_results:
    print(f"Row {result['row_index']} | Column: {result['column']} | Good Match: {'Yes' if result['is_good_match'] else 'No'}")
```
